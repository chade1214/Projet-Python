```python
# importation the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# importation the dataset
data = pd.read_csv('/content/car_prices.csv')
print(data)
```

```python
# Read the dataset
data.head()
```

```python
# verificat the information in Dataset
data.info()
```

```python
# The check for the missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Display colums with missing values
missing_values = missing_values[missing_values > 0]
print(missing_values)
```

```python
data.head()
```

```python
# Drop the column to a large number of missing
data.drop(columns =['transmission'], inplace = True)
data.isnull().sum()
```

```python
# Drop rows with missing values in 'sellingprice', 'vin' columns
data.dropna(subset=['sellingprice', 'vin'], inplace=True)
```

```python
data.isnull().sum()
```

```python
# Fill missing values in categorical columns with the most freqent value(mode)
for colmn in ['make','model','trim', 'body', 'color', 'interior']:
  data[colmn].fillna(data[colmn].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
# Fill missing values in numerical columns with the median
for column in ['odometer', 'mmr', 'condition']:
  data[column].fillna(data[column].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# Drop rows with missing values in 'saledate'
data.dropna(subset=['saledate'], inplace=True)
```

```python
data.isnull().sum()
```

```python
# generate the descriptive stactistic
descriptive_stats= data.describe()
print(descriptive_stats)
```

```python
# Filter the Dataset for old 2015
cars_2015 = data[data['year'] == 2015]
```

```python
# Count the number cars cost by make
car_count = cars_2015['make'].value_counts()
print(car_count)
```

```python
# Display the make with + numbers cost
most_sold_brand_2015 = car_count.idxmax()
most_sales_count_2015 = car_count.max()
```

```python
print(f"La marque la plus vendue en 2015 est {most_sold_brand_2015} avec {most_sales_count_2015} ventes.")
```

```python

# Plotting the horizontal bar chart
plt.figure(figsize=(12, 8))
car_count.plot(kind='barh', color='skyblue')
plt.title('Number of Cars Sold by Brand in 2015')
plt.xlabel('Number of Cars Sold')
plt.ylabel('Brand')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
car_count.plot(kind='barh', color='green')
plt.title('Number of Cars Sold by Brand in 2015')
plt.xlabel('Number of Cars Sold')
plt.ylabel('Brand')
plt.grid(True)
plt.show()
```
